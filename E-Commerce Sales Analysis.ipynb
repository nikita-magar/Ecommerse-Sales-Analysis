# E-Commerce Sales Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("Sample - Superstore.xls")
```

```python
dataset.shape
```

```python
dataset.head(2)
```

```python
dataset.isnull().sum() #to find missing values
```

# Prouct Categories

```python
plt.title("Count of Prouct Categories")
sns.countplot(x="Category",data=dataset)
plt.figure(figsize=(5,3))
plt.savefig("Count of Prouct Categories.jpg")
plt.show()
```

# Shiping Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
#To plot graph
x=dataset['Ship Mode'].value_counts().index
y=dataset['Ship Mode'].value_counts().values
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

#Shipping mode according to product category

```python
sns.countplot(x="Ship Mode",data=dataset,hue="Category")
plt.show()
```

# Cutomer Segment

```python
sns.countplot(x="Segment",data=dataset)
plt.show()
```

# Product Category

```python
sns.countplot(x="Category",data=dataset,hue="Sub-Category")
#plt.legend(loc=6)
plt.show()
```

```python
sns.countplot(x="Category",data=dataset[dataset["Category"]=="Office Supplies"],hue="Sub-Category")
plt.show()
```

# To Check number of orders accourding to year

```python
#to seprate out years from Order date
dataset["Order Year"]=dataset['Order Date'].dt.year
```

```python
dataset["Order Year"].value_counts()
```

```python
sns.countplot(x="Order Year",data=dataset)
plt.show()
```

# which catagory contains how much profit

```python
sns.barplot(x="Category",y="Profit",data=dataset,estimator="sum")
plt.show()
```
